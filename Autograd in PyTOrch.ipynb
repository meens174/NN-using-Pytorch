```python
# importing torch library
import torch
```

```python
a=torch.tensor([[1,1],[1,1]], dtype=torch.float32)
a
```

```python
b=a+5
print(b)
```

```python
c=b.mean()
print(c)
```

Automatic Differentiation

```python
a=torch.tensor([[1,1],[1,1]], dtype=torch.float32, requires_grad=True)
a
```

```python
b=a+5
c=b.mean()
print(b)
print(c)
```

```python
c.backward()
```

```python
print(a.grad)
```
